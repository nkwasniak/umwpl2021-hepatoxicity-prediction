```python
import pandas as pd

df_values = pd.read_csv('hepatotoxicity_pTD50_KlekFP_ready_set.csv', header=1)
df_values
```

```python
alts = df_values.iloc[:, 0]
alts
```

```python
alts_wo_nan=alts.dropna().reset_index().drop('index', axis=1)
alts_wo_nan
```

```python
df_after_drop=df_values.drop(df_values.columns[0], axis=1).dropna().reset_index().drop('index', axis=1)
df_after_drop
```

```python
df_after_drop.insert(0, 'TD50', alts_wo_nan)
df_after_drop
```

```python
df_after_drop.to_csv("hepatotoxicity_pTD50_KlekFP_ready_set_after_conversion.csv", index = False)
```
